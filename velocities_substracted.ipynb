```python
#Para modificar fichero csv y substraer el tiempo de los percepts para que sean equivalentes con los datos de la mirada (por el delay en el report)
import pandas as pd

# Load the CSV file
df = pd.read_csv('/Users/laurabenitosandoval/Downloads/processed_okn_velocity_data.csv')  # Updated file name

# Ensure the time_seconds column is in numeric format
df['time_seconds'] = pd.to_numeric(df['time_seconds'], errors='coerce')

# Drop rows with NaN values in time_seconds
df = df.dropna(subset=['time_seconds'])

# Shift the time_seconds column up by 28 rows when -750 ms
df['perception'] = df['perception'].shift(28)

# Drop the last 28 rows which now contain NaN values
df = df.dropna(subset=['perception'])

# Save the modified DataFrame to a new CSV file
output_file = '/Users/laurabenitosandoval/Downloads/processed_okn_velocity_data -995 ms.csv'
df.to_csv(output_file, index=False)

print(f"Modified file saved as: {output_file}")
```
